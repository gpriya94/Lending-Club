```python
data = pd.read_csv('table_0425_2.csv',header=0).rename(columns={'issued_year':'year','loan_amnt':'avgAmt','annual_inc':'avgIncome','int_rate':'avgIntRate'})
data.head()
```

```python
# Make list of years
years = data.year.value_counts().keys().sort_values().astype(str).tolist()
years
```

```python
# Make list of grades
grades = data.grade.value_counts().keys().sort_values().tolist()
grades
```

```python
columns = []
# make grid
for year in years:
    for grade in grades:
        data_by_year_and_grade = data[(data['year'] == int(year)) & (data['grade'] == grade)]
        for col_name in data_by_year_and_grade:
            # each column name is unique
            column_name = '{year}_{grade}_{header}_gapminder_grid'.format(
                year=year, grade=grade, header=col_name
            )
            a_column = Column(list(data_by_year_and_grade[col_name]), column_name)
            columns.append(a_column)
```

```python
grid = Grid(columns)
url = py.grid_ops.upload(grid, 'gapminder_grid_0425'+str(time.time()), auto_open=False)
url
```

```python
figure = {
    'data': [],
    'layout': {},
    'frames': [],
    'config': {'scrollzoom': True}
}

# fill in most of layout
figure['layout']['xaxis'] = {'range':[25000,174500],'title': 'Average Annual Income', 'gridcolor': '#FFFFFF'}
figure['layout']['yaxis'] = {'range':[1500,28000],'title': 'Average Loan Amount','gridcolor': '#FFFFFF'}
figure['layout']['hovermode'] = 'closest'
figure['layout']['plot_bgcolor'] = 'rgb(223, 232, 243)'
```

```python
sliders_dict = {
    'active': 0,
    'yanchor': 'top',
    'xanchor': 'left',
    'currentvalue': {
        'font': {'size': 20},
        'prefix': 'Year:',
        'visible': True,
        'xanchor': 'right'
    },
    'transition': {'duration': 300, 'easing': 'cubic-in-out'},
    'pad': {'b': 10, 't': 50},
    'len': 0.9,
    'x': 0.1,
    'y': 0,
    'steps': []
}
```

```python
figure['layout']['updatemenus'] = [
    {
        'buttons': [
            {
                'args': [None, {'frame': {'duration': 500, 'redraw': False},
                         'fromcurrent': True, 'transition': {'duration': 300, 'easing': 'quadratic-in-out'}}],
                'label': 'Play',
                'method': 'animate'
            },
            {
                'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                'transition': {'duration': 0}}],
                'label': 'Pause',
                'method': 'animate'
            }
        ],
        'direction': 'left',
        'pad': {'r': 10, 't': 87},
        'showactive': False,
        'type': 'buttons',
        'x': 0.1,
        'xanchor': 'right',
        'y': 0,
        'yanchor': 'top'
    }
]

custom_colors = {
    'A': 'rgb(171, 99, 250)',
    'B': 'rgb(230, 99, 250)',
    'C': 'rgb(99, 110, 250)',
    'D': 'rgb(25, 211, 243)',
    'E': 'rgb(50, 170, 255)',
    'F': 'rgb(210, 211, 243)',
    'G': 'rgb(50, 190, 115)'
}
```

```python
col_name_template = '{year}_{grade}_{header}_gapminder_grid'
```

```python
year = 2007
for grade in grades:
    data_dict = {
        'xsrc': grid.get_column_reference(col_name_template.format(
            year=year, grade=grade, header='avgIncome'
        )),
        'ysrc': grid.get_column_reference(col_name_template.format(
            year=year, grade=grade, header='avgAmt'
        )),
        'mode': 'markers',
        'textsrc': grid.get_column_reference(col_name_template.format(
            year=year, grade=grade, header='sub_grade'
        )),
        'marker': {
            'sizemode': 'area',
            'sizeref': 0.1,
            'sizesrc': grid.get_column_reference(col_name_template.format(
                 year=year, grade=grade, header='avgIntRate'
            )),
            'color': custom_colors[grade]
        },
        'name': grade
    }
    figure['data'].append(data_dict)
```

```python
for year in years:
    frame = {'data': [], 'name': str(year)}
    for grade in grades:
        data_dict = {
            'xsrc': grid.get_column_reference(col_name_template.format(
                year=year, grade=grade, header='avgIncome'
            )),
            'ysrc': grid.get_column_reference(col_name_template.format(
                year=year, grade=grade, header='avgAmt'
            )),
            'mode': 'markers',
            'textsrc': grid.get_column_reference(col_name_template.format(
                year=year, grade=grade, header='sub_grade'
            )),
            'marker': {
                'sizemode': 'area',
                'sizeref': 0.1,
                'sizesrc': grid.get_column_reference(col_name_template.format(
                 year=year, grade=grade, header='avgIntRate'
                )),
                'color': custom_colors[grade]
            },
            'name': grade
        }
        frame['data'].append(data_dict)

        
    figure['frames'].append(frame)
    slider_step = {'args': [
        [year],
        {'frame': {'duration': 300, 'redraw': False},
         'mode': 'immediate',
       'transition': {'duration': 300}}
     ],
     'label': year,
     'method': 'animate'}
    sliders_dict['steps'].append(slider_step)

figure['layout']['sliders'] = [sliders_dict]
```

```python
py.icreate_animations(figure, 'gapminder_example'+str(time.time()))
```
